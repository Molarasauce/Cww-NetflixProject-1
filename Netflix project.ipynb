# Using the for loop condition to pass an argument

```python
colors = []
for genre, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == 'Children':
        colors.append('red')
    elif row['genre']== 'Documentaries':
        colors.append('blue')
    elif row['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')
print(colors[:10])
```

```python
colors = []
for genre, row in netflix_movies_col_subset.iterrows() :
     if row['genre'] =="Children":
        colors.append('red')
     elif row['genre'] == "Documentaries" :
        colors.append('blue')
     elif row['genre'] == "Stand-Up" :
        colors.append('green')  
     else:
        colors.append('black')
print(colors[:10])
```

# Creating a scatterplot with the colors

```python
import seaborn as sns
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))
sns.scatterplot(data=netflix_movies_col_subset,x ='release_year', y ='duration', c=colors)
plt.xlabel('release_year') #x label
plt.ylabel('duration (min)')
plt.title('Movie Duration by Year of Release')
plt.show()
```
